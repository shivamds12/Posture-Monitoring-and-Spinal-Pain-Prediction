```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv('gyro_angles.csv')

# Convert timestamp to seconds from start
df['timestamp'] = df['timestamp'] - df['timestamp'].min()


# Combined visualization with thresholds
plt.figure(figsize=(15, 8))

# Calculate combined deviation magnitude
df['combined'] = np.sqrt(df['angle_x']**2 + df['angle_y']**2 + df['angle_z']**2)

# Plot all angles and combined deviation
plt.plot(df['timestamp'], df['angle_x'], alpha=0.7, label='X (Roll)')
plt.plot(df['timestamp'], df['angle_y'], alpha=0.7, label='Y (Pitch)')
plt.plot(df['timestamp'], df['angle_z'], alpha=0.7, label='Z (Yaw)')


# Threshold regions
plt.fill_between(df['timestamp'], 10, 50, where=df['combined'] > 10, 
                color='red', alpha=0.2, label='Poor Posture Zone')
plt.fill_between(df['timestamp'], -50, -10, where=df['combined'] > 10,
                color='red', alpha=0.2)

# Formatting
plt.xlabel('Time (seconds)')
plt.ylabel('Angle (degrees)')
plt.title('Posture Analysis')
plt.legend()
plt.grid(True)
plt.ylim(-50, 50)
plt.show()
```

```python
# 1. Individual time-series plots for each angle
fig, axes = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

axes[0].plot(df["timestamp"], df["angle_x"], color="blue", alpha=0.7)
axes[0].set_title("Angle X Over Time")
axes[0].set_ylabel("Angle X (degrees)")
axes[0].grid(True)

axes[1].plot(df["timestamp"], df["angle_y"], color="green", alpha=0.7)
axes[1].set_title("Angle Y Over Time")
axes[1].set_ylabel("Angle Y (degrees)")
axes[1].grid(True)

axes[2].plot(df["timestamp"], df["angle_z"], color="red", alpha=0.7)
axes[2].set_title("Angle Z Over Time")
axes[2].set_xlabel("Timestamp (s)")
axes[2].set_ylabel("Angle Z (degrees)")
axes[2].grid(True)

plt.tight_layout()
plt.show()

```
